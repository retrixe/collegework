## 3D Visualizations
The following cells add three 3D visualizations using numeric columns from the dataset.

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

# Pick first three numeric columns automatically
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) < 3:
    raise ValueError("Need at least 3 numeric columns for 3D plots.")

x_col, y_col, z_col = numeric_cols[:3]
print(f"Using columns -> X: {x_col}, Y: {y_col}, Z: {z_col}")

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[x_col], df[y_col], df[z_col], c=df[z_col], cmap='viridis', s=15, alpha=0.8)
ax.set_xlabel(x_col)
ax.set_ylabel(y_col)
ax.set_zlabel(z_col)
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# 3D Surface-style plot (triangulated)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df[x_col], df[y_col], df[z_col], cmap='plasma', linewidth=0.2, antialiased=True, alpha=0.9)
ax.set_xlabel(x_col)
ax.set_ylabel(y_col)
ax.set_zlabel(z_col)
ax.set_title('3D Tri-Surface Plot')
plt.show()
```

```python
# 3D Bar Plot (sampled for readability)
sample_df = df[[x_col, y_col, z_col]].dropna().head(200).copy()

# Normalize x and y to compact bar positions
x_vals = sample_df[x_col].to_numpy()
y_vals = sample_df[y_col].to_numpy()
z_vals = sample_df[z_col].to_numpy()

x_min, x_max = x_vals.min(), x_vals.max()
y_min, y_max = y_vals.min(), y_vals.max()

x_pos = (x_vals - x_min) / (x_max - x_min + 1e-9) * 20
y_pos = (y_vals - y_min) / (y_max - y_min + 1e-9) * 20
z_pos = np.zeros_like(z_vals)

dx = np.full_like(z_vals, 0.5, dtype=float)
dy = np.full_like(z_vals, 0.5, dtype=float)
dz = z_vals

fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, shade=True, alpha=0.8)
ax.set_xlabel(f"{x_col} (scaled)")
ax.set_ylabel(f"{y_col} (scaled)")
ax.set_zlabel(z_col)
ax.set_title('3D Bar Plot (first 200 rows)')
plt.show()
```
